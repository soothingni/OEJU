# Geoscience Data Journal 크롤러

- 2014 ~ 2019년 () 게재 논문 저자와 이메일을 크롤링하는 크롤러입니다.
- 크롤링 결과는 'Platform', 'Year', 'Volume', 'Author', 'Email', 'URL'의 6개 칼럼으로 구성된 데이터프레임으로 정리하였습니다.


```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
```

```python
root_url = 'https://rmets.onlinelibrary.wiley.com/toc/20496060/'
urls = [f"{root_url}{i}/{i-2013}/" for i in range(2014, 2020)]
urls = [url + str(k) for k in range(1, 3) for url in urls]
```

```python
urls.sort()
```

```python
urls[3].split('/')
```

```python
df = pd.DataFrame(columns = ['Platform', 'Year', 'Volume', 'Author', 'Email', 'URL'])

driver = webdriver.Chrome()

for url in urls:
    driver.get(url)
    article_list = driver.find_elements_by_class_name('issue-item__title.visitable')
    article_urls = [article.get_attribute('href') for article in article_list][1:]    #첫번째 글은 issue 소개글(저자 정보 없음)
    for k in range(len(article_urls)):
        driver.get(article_urls[k])
        button = driver.find_element_by_xpath('//*[@id="a1_Ctrl"]/span')
        author = button.text
        button.click()
        email = driver.find_element_by_class_name('corr-email').get_attribute('href').replace('mailto:', '')
        
        
        df = df.append({'Platform': 'Geoscience', 'Year': url.split('/')[-3], 'Volume': url.split('/')[-2], 'Author': author, 'Email': email, 'URL': article_urls[k] }, ignore_index=True)
        time.sleep(1)
        
print('finished')
```
